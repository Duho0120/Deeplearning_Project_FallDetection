```python
import shutil
from pathlib import Path

# ì†ŒìŠ¤ ê²½ë¡œ (ì›ë³¸ ì´ë¯¸ì§€ê°€ ìˆëŠ” ê²½ë¡œ - êµ¬ë„_C2ë¡œ ëë‚˜ëŠ” í´ë”ë“¤ì„ ì°¾ì„ ê²½ë¡œ)
source_base_path = Path(r"C:\Users\ASUS\Desktop\ì œë¡œë² ì´ìŠ¤\ë”¥ëŸ¬ë‹ í”„ë¡œì íŠ¸\ë‚™ìƒì‚¬ê³  ìœ„í—˜ë™ì‘ ì˜ìƒ-ì„¼ì„œ ìŒ ë°ì´í„°_ë³‘ì›,í›„ë©´ë‚™ìƒ\3.ê°œë°©ë°ì´í„°\1.ë°ì´í„°\Training\01.ì›ì²œë°ì´í„°\TS\ì´ë¯¸ì§€\Y\SY")

# ëŒ€ìƒ ê²½ë¡œ (ëª¨ì€ì´ë¯¸ì§€\êµ¬ë„_C2 í´ë”)
current_project_path = Path.cwd()
destination_path = current_project_path / "ëª¨ì€ì´ë¯¸ì§€" / "êµ¬ë„_C2"

print(f"ì†ŒìŠ¤ ê¸°ë³¸ ê²½ë¡œ: {source_base_path}")
print(f"ëŒ€ìƒ ê²½ë¡œ: {destination_path}")
print(f"ì†ŒìŠ¤ ê²½ë¡œ ì¡´ì¬ ì—¬ë¶€: {source_base_path.exists()}\n")

# ì†ŒìŠ¤ ê²½ë¡œê°€ ì¡´ì¬í•˜ëŠ”ì§€ í™•ì¸
if not source_base_path.exists():
    print(f"âŒ ì˜¤ë¥˜: ì†ŒìŠ¤ ê²½ë¡œê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {source_base_path}")
else:
    # ëŒ€ìƒ í´ë” ìƒì„±
    destination_path.mkdir(parents=True, exist_ok=True)
    print(f"âœ… ëŒ€ìƒ í´ë” ìƒì„±/í™•ì¸ ì™„ë£Œ: {destination_path}\n")
    
    # êµ¬ë„_C2ë¡œ ëë‚˜ëŠ” í´ë” ì°¾ê¸°
    c2_folders = [f for f in source_base_path.iterdir() if f.is_dir() and f.name.endswith('C2')]
    
    if not c2_folders:
        print(f"âš ï¸  ê²½ê³ : êµ¬ë„_C2ë¡œ ëë‚˜ëŠ” í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    else:
        print(f"ğŸ“ ë°œê²¬ëœ êµ¬ë„_C2 í´ë” ìˆ˜: {len(c2_folders)}ê°œ\n")
        
        total_copied = 0
        total_skipped = 0
        
        # ê° í´ë”ì—ì„œ ì´ë¯¸ì§€ íŒŒì¼ ë³µì‚¬
        for folder in c2_folders:
            print(f"ğŸ“‚ ì²˜ë¦¬ ì¤‘: {folder.name}")
            
            # ì´ë¯¸ì§€ íŒŒì¼ ì°¾ê¸° (jpg, jpeg, png ë“±)
            image_extensions = {'.jpg', '.jpeg', '.JPG', '.JPEG', '.png', '.PNG'}
            image_files = [f for f in folder.rglob('*') if f.is_file() and f.suffix in image_extensions]
            
            if not image_files:
                print(f"   âš ï¸  ì´ë¯¸ì§€ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.\n")
                continue
            
            copied_in_folder = 0
            skipped_in_folder = 0
            
            for img_file in image_files:
                dest_file = destination_path / img_file.name
                
                if dest_file.exists():
                    skipped_in_folder += 1
                else:
                    try:
                        shutil.copy2(img_file, dest_file)
                        copied_in_folder += 1
                    except Exception as e:
                        print(f"   âŒ ì˜¤ë¥˜ ({img_file.name}): {e}")
            
            total_copied += copied_in_folder
            total_skipped += skipped_in_folder
            print(f"   âœ… ë³µì‚¬: {copied_in_folder}ê°œ, ê±´ë„ˆëœ€: {skipped_in_folder}ê°œ\n")
        
        print(f"\nâœ… ì „ì²´ ë³µì‚¬ ì™„ë£Œ!")
        print(f"   - ì´ ë³µì‚¬ëœ íŒŒì¼: {total_copied}ê°œ")
        print(f"   - ì´ ê±´ë„ˆë›´ íŒŒì¼: {total_skipped}ê°œ")
        print(f"   - ëŒ€ìƒ ê²½ë¡œ: {destination_path}")
        
        # ìµœì¢… íŒŒì¼ ê°œìˆ˜ í™•ì¸
        final_count = sum(1 for _ in destination_path.glob('*') if _.is_file())
        print(f"   - ëŒ€ìƒ í´ë”ì˜ ì´ íŒŒì¼ ìˆ˜: {final_count}ê°œ")
```
